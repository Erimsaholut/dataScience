# Finding Relationships

A great aspect of the Pandas module is the corr() method.

The corr() method calculates the relationship between each column in your data set.

The examples in this page uses a CSV file called: 'data.csv'.

```python
import pandas as pd

df = pd.read_csv("data.csv")

print(df.corr())
```

Note: The corr() method ignores "not numeric" columns. 

Evet, eğer bir sütunda sadece tek bir hücrede bile numeric olmayan bir değer (örneğin 5 yerine "five") varsa, corr() yöntemi o sütunu tamamen analiz dışı bırakır. Yani corr(), sütunun tamamını sayısal değilmiş gibi değerlendirir ve o sütundaki numeric olmayan tek bir veri bile sütunun tamamının korelasyon hesaplamasından çıkarılmasına neden olur.

Bu tür durumlarda, sütunun sadece sayısal verilerini koruyup diğer değerleri temizlemek veya dönüştürmek için veri ön işleme yapılması önerilir. Örneğin, yanlış girilmiş five değerini 5 olarak düzeltebilir ya da pd.to_numeric(errors='coerce') gibi bir yöntemle numeric olmayan verileri NaN'e dönüştürebilirsiniz.

## Result Explained

The Result of the corr() method is a table with a lot of numbers that represents how well the relationship is between two columns.

The number varies from -1 to 1.

1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

#### What is a good correlation?
It depends on the use, but I think it is safe to say you have to have at least 0.6 (or -0.6) to call it a good correlation.

matplotlib.pyplot, matplotlib kütüphanesinin en yaygın kullanılan alt kütüphanelerinden biridir ve veri görselleştirmesi için güçlü bir araç sağlar. Bunun dışında, matplotlib kütüphanesinin başka alt modülleri de vardır. İşte en yaygın kullanılan matplotlib alt modülleri ve görevleri:
1. matplotlib.pyplot

    Görev: Çizimler ve grafikler oluşturmak için kullanılır.
    Kullanımı: Grafik türleri, etiketleme, eksen ayarları, renkler, stiller vb. gibi temel görselleştirme araçlarını sağlar. Örneğin, çizgi grafikler, çubuk grafikler, histogramlar ve dağılım grafiklerini oluşturabilirsiniz.
    Örnek:

    import matplotlib.pyplot as plt
    plt.plot([1, 2, 3], [4, 5, 6])
    plt.show()

2. matplotlib.figure

    Görev: figure (şekil) nesnelerini yönetir ve bir grafik veya grafik seti için boş bir tuval sağlar.
    Kullanımı: Tek bir figür oluşturmak, birden fazla eksen (subplots) eklemek gibi amaçlarla kullanılır.
    Örnek:

    from matplotlib.figure import Figure
    fig = Figure(figsize=(5, 4))

3. matplotlib.axes

    Görev: axes (eksen) nesnelerini yönetir ve verilerin görüntüleneceği alanı temsil eder.
    Kullanımı: Grafik üzerinde çizilecek olan eksenlerin ayarlarını yapar ve birçok özelliği kontrol eder (x ve y eksenleri, başlık, vb.).
    Örnek:

    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    ax.plot([1, 2, 3], [4, 5, 6])

4. matplotlib.animation

    Görev: Animasyonlu grafikler oluşturmak için kullanılır.
    Kullanımı: Verilerin zamanla değişimini göstermek için animasyon oluşturma araçları sağlar.
    Örnek:

    from matplotlib.animation import FuncAnimation

5. matplotlib.colors

    Görev: Renk haritaları, renk ölçekleri ve renklerle ilgili diğer işlemler için kullanılır.
    Kullanımı: Belirli bir renk haritası tanımlamak veya renklere dair özelleştirme yapmak için faydalıdır.
    Örnek:

    from matplotlib import colors
    cmap = colors.ListedColormap(['red', 'green', 'blue'])

6. matplotlib.cm

    Görev: Renk haritalarını (colormaps) yönetir.
    Kullanımı: Renk gradyanları ve özel renk haritalarıyla çalışmak için kullanılır.
    Örnek:

    import matplotlib.pyplot as plt
    plt.imshow(data, cmap=plt.cm.viridis)

7. matplotlib.gridspec

    Görev: Grafik düzenleri oluşturmak için kullanılır.
    Kullanımı: Grafik içindeki alt grafiklerin (subplots) düzenini kontrol etmek için faydalıdır.
    Örnek:

    from matplotlib import gridspec
    gs = gridspec.GridSpec(2, 2)

8. matplotlib.dates

    Görev: Zaman serisi verilerini çizmek için kullanılan eksenlerin ayarlanmasını sağlar.
    Kullanımı: Tarih ve zaman formatları üzerinde özelleştirme yapmak için kullanılır.
    Örnek:

    import matplotlib.pyplot as plt
    import matplotlib.dates as mdates

9. matplotlib.ticker

    Görev: Grafiklerin eksenlerinde gösterilen değerleri ve çizgileri (ticks) özelleştirmek için kullanılır.
    Kullanımı: X ve Y ekseninde daha gelişmiş etiketlemeler yapmanızı sağlar.
    Örnek:

    from matplotlib import ticker

Bu alt modüller, matplotlib kütüphanesinin esnekliğini artırır ve görselleştirme ihtiyaçlarına göre farklı alanlarda özelleştirme yapmanıza olanak tanır.

## Histogram

Use the kind argument to specify that you want a histogram:

kind = 'hist'

A histogram needs only one column.

A histogram shows us the frequency of each interval, e.g. how many workouts lasted between 50 and 60 minutes?

In the example below we will use the "Duration" column to create the histogram:

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

df["Duration"].plot(kind="hist")

plt.show()
```
